```python
import pandas as pd
from matplotlib import pyplot as plt

# Read CSV into pandas
data = pd.read_csv("diabetes.csv")
data.head()
df = pd.DataFrame(data)
Age = df['Age'].head(12)
BloodPressure = df['BloodPressure'].head(12)

# Figure Size
fig = plt.figure(figsize =(10, 7))

# Horizontal Bar Plot
plt.bar(Age[0:10], BloodPressure[0:10])

# Show Plot
plt.show()
```
